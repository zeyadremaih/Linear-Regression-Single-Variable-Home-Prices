```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price)
```

```python
#Suitable for linear regression model
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
X_test=pd.read_csv('X_test.csv',index_col=None)
```

```python
X_test
```

```python
predictions=reg.predict(X_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=m*x+b
y=(reg.coef_)*(X_test)+(reg.intercept_)
```

```python
result=pd.DataFrame(X_test)
```

```python
result['Prices Predictions']=predictions
```

```python
result.to_csv('result',index=False)
```
